```python

import random

def volados(n):
    """Simula n volados y calcula la probabilidad de A y S."""
    x = []  # Lista para almacenar los resultados
    conteo_A = 0  # Contador de "A"
    conteo_S = 0  # Contador de "S"

    for _ in range(n):  # Genera una lista de n valores aleatorios
        num = random.random()  # Genera un número aleatorio entre 0 y 1

        if num < 0.5:
            x.append("A")  # Si es menor que 0.5, agrega "A"
            conteo_A += 1  # Suma 1 al contador de "A"
        else:
            x.append("S")  # Si es 0.5 o mayor, agrega "S"
            conteo_S += 1  # Suma 1 al contador de "S"

    # Calcular probabilidades
    prob_A = conteo_A / n
    prob_S = conteo_S / n

    return x, prob_A, prob_S  # Retorna la lista y las probabilidades

# Llamar a la función
n = 25  # Número de volados
resultado, probabilidad_A, probabilidad_S = volados(n)

# Imprimir resultados
print("Resultados:", resultado[:50], "...")  # Muestra solo los primeros 50 resultados
print(f"Probabilidad de A: {probabilidad_A:.2%}")
print(f"Probabilidad de S: {probabilidad_S:.2%}")

def experimento(m):
 p=[]

 for i in range(m):
  a=volado(i)
  p.append(a)
 return p

# Llamar a la función con m experimentos
m = 10
resultado = experimento(m)

# Imprimir resultados
print("n  | Probabilidad de A | Probabilidad de S")
print("-----------------------------------------")
for n, prob_A, prob_S in resultado:
    print(f"{n:<3} | {prob_A:.2%}
```

```python

import random
import matplotlib.pyplot as plt

def volados(n):
    """Simula n volados y calcula la probabilidad de A y S."""
    conteo_A = 0

    for _ in range(n):
        num = random.random()
        if num < 0.5:
            conteo_A += 1

    prob_A = conteo_A / n if n > 0 else 0
    return prob_A

def experimento(m):
    """Ejecuta m experimentos con volados y almacena las probabilidades."""
    valores_n = []
    probabilidades_A = []

    for n in range(1, m + 1):
        prob_A = volados(n)
        valores_n.append(n)
        probabilidades_A.append(prob_A)

    return valores_n, probabilidades_A

# Número de experimentos
m = 1000
n_valores, prob_A_valores = experimento(m)

# Graficar la evolución de la probabilidad
plt.figure(figsize=(10, 5))
plt.plot(n_valores, prob_A_valores, label="Probabilidad de A", color="blue")
plt.axhline(y=0.5, color="red", linestyle="--", label="Valor teórico (50%)")
plt.xlabel("Número de volados (n)")
plt.ylabel("Probabilidad de A")
plt.title("Convergencia de la probabilidad de obtener A en un volado")
plt.legend()
plt.grid(True)
plt.show()
```
