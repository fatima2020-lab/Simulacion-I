```python
# Alumna: Martinez Arenas Fatima Areli

####################### PRUEBA DE CORRIDAS (RACHAS) ############################


# Librerias a utilizar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from tabulate import tabulate # Para mostrar tablas formateadas

# Definir la muestra de datos
muestra = [0.347, 0.993, 0.674, 0.426, 0.46, 0.189, 0.112, 0.37, 0.909, 0.178,
           0.832, 0.371, 0.628, 0.054, 0.224, 0.753, 0.191, 0.314, 0.764, 0.516,
           0.966, 0.729, 0.055, 0.022, 0.99, 0.73, 0.584, 0.731, 0.999, 0.437,
           0.472, 0.067, 0.494, 0.742, 0.786, 0.797, 0.347, 0.742, 0.303, 0.393,
           0.797, 0.189, 0.494, 0.674, 0.393, 0.292, 0.426, 0.213, 0.718, 0.268,
           0.101, 0.977, 0.235, 0.898, 0.461, 0.876, 0.057, 0.472, 0.933, 0.123,
           0.696, 0.843, 0.178, 0.641, 0.011, 0.707, 0.819, 0.641, 0.056, 0.945,
           0.966, 0.562, 0.775, 0.674, 0.977, 0.562, 0.303, 0.944, 0.415, 0.527,
           0.404, 0.549, 0.797, 0.821, 0.246, 0.821, 0.404, 0.28, 0.819, 0.459,
           0.603, 0.992, 0.252, 0.19, 0.881, 0.112, 0.64, 0.663, 0.444, 0.652 ]


# Convertir a numpy array y transponer en columna
muestra_transpuesta = np.array(muestra).reshape(-1, 1)

# Imprimir la lista en formato de columna
#for valor in muestra_transpuesta:
    #print(valor[0])

# Primer paso: Determinar una secuencia S de unos y ceros generando un
 # - 0 si xi < xi-1
 # - 1 si xi >= xi-1
S = []
for i in range(1, len(muestra_transpuesta)):
  if muestra_transpuesta[i] >= muestra_transpuesta[i-1]:
    S.append(1)
  else:
    S.append(0)

# Convertir a numpy array y transponer en columna a S
S_transpuesta = np.array(S).reshape(-1,1)

# Añadir un valor para la primera fila de la secuencia (ya que S tiene una longitus de n-1)
xi = muestra_transpuesta[1:].reshape(-1,1)

# Imprimir la lista en formato columna
#for valors in S_transpuesta:
  #print(valors[0])

# Paso 2: Hallar el número de corridas (rachas) observadas C_0. Una corridad
# viene dada por grupos de "1" o "0"

def contar_rachas(S):
  """
    Cuenta el número de corridas (rachas) en la secuencia S. Una corrida es una secuencia
    de valores consecutivos iguales (ya sea 1 o 0).

    :param S: Lista de 0s y 1s que representa la secuencia de corridas.
    :return: Número de corridas (rachas).
    """
  rachas = 1  # Empezar con la primera racha
  for i in range(1, len(S)):
    if S[i] != S[i-1]:   # SI hay cambio de valor (de 1 a 0 o de 0 a 1)
     rachas += 1
  return rachas

# Llamar a la función para contar las corridas
Co = contar_rachas(S)

# Paso 3: Calcular el valor esperado del numero de corridas
mu = ((2 * len(muestra_transpuesta))-1)/3

# Paso 4: Calcular la varianza del número de corridas
sigma2 = ((16 * len(muestra_transpuesta))-29)/90

#Paso 5: Calcular a z0 (El valor estadistico)
z0 = (Co-mu)/np.sqrt(sigma2)

#########################Crear tabla de reusultados#############################

# Convertir las listas en una tabla de formato adecuado
tabla_resultados = np.hstack((xi, S_transpuesta))

headers = ["xi", "Corridas"]
print("\nTabla de corridas observadas")
print(tabulate(tabla_resultados, headers=headers, tablefmt="grid"))

####################### INTERVALO DE ACEPTACIÓN ###############################

# El intervalo de aceptación esta dado por  -z_{\alpha/2} <= z0 <= z_{\alpha/2
# Definir itervalo de aceptación
def valor_Z(alpha):
  """
    Calcula el valor crítico Z para un nivel de significancia alpha dado, utilizando la distribución normal estándar.

    :param alpha: Nivel de significancia (por ejemplo, 0.05, 0.01).
    :return: El valor crítico Z correspondiente a alpha.
    """
  # Calcular Z_alpha/2 utilizando distribución normal estandar
  return stats.norm.ppf(1 - alpha/2)

# Definir los niveles de confianza y sus valores de alpha
nivel_aceptacion = { '90%' : 0.10, '95%' : 0.05, '99%' : 0.01}

for nivel, alpha in nivel_aceptacion.items():
  Z_alpha_2 = valor_Z(alpha)

############################### RESULTADOS ###################################

print(f"Número de corridas observadas (Co): {round(Co,3)}")
print(f" Valor esperado de corridas (mu): {round(mu, 3)}")
print(f"Varianza del número de corridas (sigma2): {round(sigma2, 3)}")
print(f"Valor estadistico (z0): {round(z0, 3)}")

# Solicitar al usuario que seleccione un nivel de aceptación
print("Seleccione el nivel de aceptación:")
for nivel in nivel_aceptacion:
  print(f" - {nivel}")

nivel_seleccionado = input("Ingrese el nivel de confianza ")

# Validar que el nivel ingresado sea válido
if nivel_seleccionado in nivel_aceptacion:
  alpha = nivel_aceptacion[nivel_seleccionado]
  Z_alpha_2 = valor_Z(alpha)
  print(f"Para un nivel de aceptación del {nivel_seleccionado}, Z_alpha/2 es: {Z_alpha_2:.3f}")
else:
  print("Nivel de aceptación no válido. Por favor ingrese 90%, 95%, 99%.")

# Definir el intervalo de aceptación
if -Z_alpha_2 <= z0 <= Z_alpha_2:
    print("\nNo se rechaza la hpótesis nula. Las categorias tienen una distribución similar.")

else:
    # Si el valor de p es mayor que alpha, no se rechaza la hipótesis nula, por lo que las categorías tienen una distibución similar y los datos son independientes.
    print("\nSe rechaza la hipótesis nula. Las categorías tienen una distribución diferente.")





```
