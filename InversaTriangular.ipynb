**METODO DE LA TRANSFORMADA INVERSA**


*   Distribución Triangular



```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random as rd
```

Proporcionar los parametros:  **a, b y c**

```python
a= 1
b = 3
c = 5
```

Definir la función de  **F(x) inversa**

\begin{cases}
\frac{(x - a)^2}{(b - a)(c - a)} & \text{si } a \leq x \leq b \\
1 - \frac{(c - x)^2}{(c - a)(c - b)} & \text{si } b < x \leq c
\end{cases}


```python
def inversa(n):
  x = []
  for i in range(n):
    R = rd.random()
    if  R < 0.5:
      x.append(a + math.sqrt(R*(b-a)*(c-a)))
    else:
      x.append(c - math.sqrt((1-R)*(c-a)*(c-b)))
  return x
```

```python
u = inversa(10000)
print(u)
```

Graficar la distribucion triangular


```python
plt.figure(figsize=(10,6))
plt.hist(u, bins=100, density=True, alpha=0.6, color='lightblue', edgecolor='black')
plt.show()
```

Calcular la media


```python
media = np.mean(u)
print(media)
```

Calcular la Varianza


```python
varianza = np.var(u)
print(varianza)
```
