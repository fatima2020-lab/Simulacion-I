```python
# Alumna: Martinez Arenas Fatima Areli

## Pruba De Kolmogorov-Smironov

# Librerias necesarias
import numpy as np  # Librería numpy, manipular arrays y matrices, así como realizas operaciones
import pandas as pd # Librería para la manipulación y el análisis de datos, tablas, series.
from scipy.stats import uniform # Librería para la estadística
from tabulate import tabulate # Librería para mostrar tabla de manera estructurada
import math
# Lista de Datos de Muestra
muestra = [0.78961, 0.05230, 0.10699, 0.55877, 0.14151, 0.76086, 0.12079, 0.27738, 0.65726,
    0.79269, 0.80548, 0.82654, 0.29453, 0.20852, 0.42989, 0.58518, 0.98611, 0.34488,
    0.34358, 0.11537, 0.89898, 0.57880, 0.67621, 0.05010, 0.00121, 0.28269, 0.73059,
    0.70119, 0.18284, 0.49962, 0.38618, 0.76910, 0.68334, 0.55170, 0.10850, 0.79982,
    0.45679, 0.21631, 0.87616, 0.55743, 0.58962, 0.33216, 0.03185, 0.61168, 0.09264,
    0.69623, 0.17028, 0.05475, 0.91512, 0.76262, 0.29931, 0.30861, 0.83358, 0.51781,
    0.03272, 0.57410, 0.26593, 0.85903, 0.43398, 0.35286, 0.24000, 0.65559, 0.38507,
    0.90829, 0.94187, 0.93655, 0.88809, 0.81772, 0.36982, 0.19904, 0.54325, 0.62400,
    0.09133, 0.41678, 0.33954, 0.58244, 0.85853, 0.88752, 0.33729, 0.15506, 0.23949,
    0.53559, 0.33381, 0.49383, 0.75103, 0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724, 0.22287, 0.07281, 0.64183, 0.44267,
    0.72102]

# Paso 1: Ordenar de menor a mayor los numeros aleatorios de la muestra
muestra_ordenada = np.sort(muestra)
n = len(muestra)  # Número de datos
resultados=[]

# Paso 2: Calcular a D más (Distancia máxima hacia arriba) y  D menos (Distancia máxima hacia abajo)
for i in range(n):
  D_mas = abs((i/n) - muestra_ordenada[i] )
  D_menos = abs(muestra_ordenada[i] - (i+1) /n)
  resultados.append([i + 1, muestra_ordenada[i], D_mas, D_menos])

# Calcular el valor máximos de D+ y D-
D_mas_max = max([row[2] for row in resultados])
D_menos_max = max([row[3] for row in resultados])

# Paso 4: Calcular a D (Estadístico de pruba de Kolmogorov Smirnov)
D = max(D_mas_max, D_menos_max)

# Paso 5: Calcular el valor critico
ks_valor_critico = math.sqrt(-0.5*math.log(0.05/2) / n)

# Lista de resultados para la tabla de Kolmogorov-Smirnov
ks_resultados = []

# Llenar lista con resultados
for row in resultados:
  ks_resultados.append([row[0], row[1], row[2], row[3]])

# Ordenar los resultados por el valor de i
ks_resultados.sort(key=lambda x: x[0])

# Crear tabla de resultados
ks_tabla=[]

# Se llena con los resultados:
# - El numero de i
# - El valor xi
# - El valor de D+
# - El valor de D-

for i, xi, D_mas, D_menos in ks_resultados:
  ks_tabla.append([i, xi, D_mas, D_menos])

# Mostrar los reultado en una tabla
def imprimir_ks_resultados(ks_tabla):
    """
    Imprime una tabla con los resultados de la prueba Kolmogorov-Smirnov de manera estructurada.

    :param tabla_resultados_ks: Lista con los resultados de la prueba KS
    """
    headers = ["i", "xi", "D+", "D-"]
    print(tabulate(ks_tabla, headers=headers, tablefmt="grid"))

# Llamar al función para mostra la tabla de KS
imprimir_ks_resultados(ks_tabla)

# Mostrar Resultados
print(f"\nEstadístico KS: {D}")
print(f"\nValor críticode KS: {ks_valor_critico}")

# Paso 6: Comparar el estadístico KS con el valor crítico para determinar si se rechaza la hipótesis nula
if D > ks_valor_critico:
    print("\nSe rechaza la hipótesis nula. La muestra no sigue la distribución uniforme.")
else:
    print("\nNo se rechaza la hipótesis nula. La muestra sigue la distribución uniforme.")

```
