Generador congruencial del cuadrado medio

```python
# Alumna: Martinez Arenas Fatima Areli
# Generador Congruencial del cuadrado medio

from tabulate import tabulate

def cuadrado_medio(x0, cantidad):
    """
    Implementa el método del cuadrado medio para generar números pseudoaleatorios.

    Parámetros:
    x0 (int): Semilla inicial.
    cantidad (int): Cantidad de números pseudoaleatorios a generar.

    Retorna:
    ciclo_vida (list): Lista de listas con la iteración, el número generado y su versión normalizada.
    numeros_aleatorios (list): Lista de los números generados (no utilizada en este código).
    """
    numeros_aleatorios = []
    num = x0
    ciclo_vida = []

    for i in range(1, cantidad + 1):
        cuadrado = num ** 2 # Semilla al cuadrado

        # Convierte el cuadrado a una cadena y extrae la parte central
        cuadrado_cad = str(cuadrado).zfill(8)
        mitad = len(cuadrado_cad) // 2
        num = int(cuadrado_cad[mitad-2:mitad+2])  # Toma 4 dígitos del centro

        num_normalizado = num / 10000  # Asumimos una normalización entre 0 y 1

        # Agrega el número generado, su versión normalizada y la iteración
        ciclo_vida.append([i, num, num_normalizado])

    return ciclo_vida, numeros_aleatorios

# Función para imprimir el ciclo de vida de la semilla y los números generados
def imprimir_ciclo_vida(ciclo_vida):
    """
    Imprime en formato tabular la lista de números generados.

    Parámetros:
    ciclo_vida (list): Lista con la información del ciclo de vida de los números generados.
    """
    # Imprimir la tabla usando tabulate
    headers = ["Iteración", "Número", "Número Normalizado"]
    print(tabulate(ciclo_vida, headers=headers, tablefmt="grid"))



# Variables
x0 = 1978
cantidad = 50
ciclo_vida, resultados = cuadrado_medio(x0, cantidad)

# Imprimir la tabla con los resultados
print("Ciclo de vida de la semilla:")
imprimir_ciclo_vida(ciclo_vida)
```
