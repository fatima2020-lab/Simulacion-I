Dos personas han quedado citas en un determinado lugar. Ambas tienen la misma
probabilidad de llegar al lugar de la cita en el intervalo de tiempo [0, 30] (minutos).
Determínese la probabilidad de que el tiempo que una persona espere a la otra sea
como máximo 10 minutos.

**Solución.**


Paso 1: Identificar las variables.

*   $X$: Tiempo de llegada de la Persona 1, en minutos.
*  $Y$: Tiempo de llegada de la Persona 2, en minutos.
*  $D=|X-Y|$: Diferencia absoluta entre los tiempos de llegada.

**Paso 2: Determinar la distribución de probabilidad.**

Tanto $X$ como $Y$ se distribuyen de forma **uniforme continua** en el intervalo $[0,30]$, es decir:

$X∼U(0,30),$  $Y∼U(0,30)$

**Paso 3: Modelar las variables aleatorias**

Se modelarán $X$ y $Y$ usando una función generada de npumeros aleatorios con distribución uniforme.

Se generarán pares de valores $(X_i, Y_i)$, y se calculará:

$D_i=|X_i-Y_i|$

**Paso 4: Definir el modelo y los objetivos de la simulación.**

* Modelo: Para cada iteración se simula una pareja de llegadas aleatorias y se evalúa si la diferencia $D\leq 10$.
* Objetivo: Estimar la probabilidad de que la espera entre personas sea como máximo 10 minutos, es decir:
$P(|X-Y|\leq10)$

**Paso 5: Diseñar el experimento.**
*  Número de repeticiones: N= 100,000 simulaciones.

* Para cada simulación:
  *  Generar $X_i, Y_i∼U(0,30)$
  *  Calclar $D_i=|X_i-Y_i|$
  *  Verificar si $D_i\leq 10$

**Paso 6: Repetir M veces el experimento**

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de simulaciones
N = 100000

# Paso 6: Generar las llegadas
X = np.random.uniform(0, 30, N)  # Llegadas de Persona 1
Y = np.random.uniform(0, 30, N)  # Llegadas de Persona 2

# Diferencias absolutas
D = np.abs(X - Y)

# Contar cuántas diferencias cumplen la condición
eventos_favorables = D <= 10

# Probabilidad estimada
probabilidad = np.mean(eventos_favorables)
print(f"Probabilidad estimada por simulación: {probabilidad:.4f}")
```

**Paso 7: Graficar resultados.**

```python
# Graficar puntos (azul = espera ≤ 10, rojo = espera > 10)
plt.scatter(X, Y, c=eventos_favorables, cmap='bwr', s=1, alpha=0.3)

plt.xlabel('Persona 1')
plt.ylabel('Persona 2')
plt.title('Llegadas y espera ≤ 10 minutos')
plt.grid(True)
plt.show()
```
